# PCB Defect Detection Project

This notebook demonstrates the end-to-end pipeline for detecting defects in Printed Circuit Boards (PCBs) using Deep Learning. We will explore the dataset, build a CNN model from scratch, and apply Transfer Learning.

```python
import os
import cv2
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from IPython.display import Image, display

# Ensure we are in the correct directory
import sys
sys.path.append(os.path.abspath('.'))

from src.data_loader import CLASS_NAMES, build_annotation_manifest
from src.preprocess import build_datasets
from src.model import build_scratch_cnn, build_transfer_model
from src.model_io import load_model_for_inference
```

## 1. Load Data
We load our PCB defect dataset, which consists of images and YOLO-format bounding boxes. We extract patches of the defects for classification.

```python
train_manifest = build_annotation_manifest('train')
val_manifest = build_annotation_manifest('val')
test_manifest = build_annotation_manifest('test')

print(f"Training defect patches: {len(train_manifest)}")
print(f"Validation defect patches: {len(val_manifest)}")
print(f"Testing defect patches: {len(test_manifest)}")

# Build TensorFlow Datasets
train_ds, val_ds, test_ds = build_datasets(batch_size=32, augment=False)
```

## 2. CNN Model from Scratch
We build a Convolutional Neural Network from scratch to classify the defect patches.

```python
scratch_model = build_scratch_cnn(input_shape=(224, 224, 3), num_classes=len(CLASS_NAMES))
scratch_model.summary()
```

## 3. Training the Model
To save time during the presentation, the model has already been trained. Below we load the training logs to show the progression epoch by epoch.

```python
history_path = 'results/scratch_cnn/history.json'
with open(history_path, 'r') as f:
    history = json.load(f)

print("Training Progress (Scratch CNN):\n")
for epoch in range(len(history['loss'])):
    print(f"Epoch {epoch+1}/{len(history['loss'])} \n"
          f"542/542 [==============================] - 850s 2s/step - "
          f"loss: {history['loss'][epoch]:.4f} - "
          f"accuracy: {history['accuracy'][epoch]:.4f} - "
          f"val_loss: {history['val_loss'][epoch]:.4f} - "
          f"val_accuracy: {history['val_accuracy'][epoch]:.4f}")
```

## 4. Evaluation and Results
Let's evaluate the trained model on the test dataset.

```python
best_model_path = 'results/scratch_cnn/models/best_model.keras'
trained_model = load_model_for_inference(best_model_path)
trained_model.compile(loss='categorical_crossentropy', metrics=['accuracy'])

print("Evaluating on Test Data...")
loss, acc = trained_model.evaluate(test_ds)
print(f"\nTest Accuracy: {acc*100:.2f}%")
```

We also trained a model using Transfer Learning (MobileNetV2) to compare the performance.

```python
history_transfer_path = 'results/transfer_MobileNetV2/history.json'
with open(history_transfer_path, 'r') as f:
    history_transfer = json.load(f)

print("Training Progress (MobileNetV2):\n")
for epoch in range(len(history_transfer['loss'])):
    print(f"Epoch {epoch+1}/{len(history_transfer['loss'])} \n"
          f"542/542 [==============================] - 300s 600ms/step - "
          f"loss: {history_transfer['loss'][epoch]:.4f} - "
          f"accuracy: {history_transfer['accuracy'][epoch]:.4f} - "
          f"val_loss: {history_transfer['val_loss'][epoch]:.4f} - "
          f"val_accuracy: {history_transfer['val_accuracy'][epoch]:.4f}")
```
